## Importing libraries 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import math
```

Create a list with normal distribution of 500 candy bars with mean 18.78 and standard deviation of 0.58.

```python
pop = np.random.normal(18.79,0.58,500)
```

Population mean

```python
pop.mean()
```

Population standard deviation

```python
pop.std()
```

Population histogram

```python
plt.hist(pop)
plt.title('Population Histogram')
plt.xlabel('Individual Candy Bar weights')
```

Create 500 samples of size 10 from population and display sampling distribution. The distribution looks normal and majority of the samples are in the curve

```python
np.random.seed(10)

point_estimates = []         # Make empty list to hold point estimates

for x in range(500):         # Generate 200 samples
    sample = np.random.choice(a= pop, size=10)
    point_estimates.append( sample.mean() )
    
f, ax = plt.subplots(figsize=(12, 10))
sns.distplot( point_estimates, hist=False, rug=True); # Plot sample mean density
```

The samples histogram. It looks close to the population histogram but it looks like more data is centered around the mean. By the central limit theorem, the distribution of many sample means, known as a sampling distribution, will be normally distributed.

```python
plt.hist(point_estimates)
plt.title('Sample Histogram')
plt.xlabel('Sample Mean Weights of Candy Bars')
```

The samples mean

```python
np.array(point_estimates).mean()
```

Sample standard deviation

```python
np.array(point_estimates).std()
```

The difference between the population mean - sample means, the difference appears not to be that significant. The reason why the sample mean is close to the population mean because sample mean is an unbiased estimate of the population mean. This means that the sample mean is not systematically smaller or larger than the population mean.

```python
pop.mean() - np.array(point_estimates).mean()
```

The difference between population standard deviation and sample deviation, the difference appears to be big. This is because the sample standard deviation is divided by the square root of n. This will make the sample standard deviation smaller when compared to the population standard deviation.

```python
pop.std() - np.array(point_estimates).std()
```

# Confidence Interval
A confidence interval is a range of values above and below a point estimate that captures the true population parameter at some predetermined confidence level

```python

z_critical = stats.norm.ppf(q = 0.975)  # Get the z-critical value*
sample_size = 500
sample_mean = np.array(point_estimates).mean()

pop_stdev = pop.std()  # Get the population standard deviation

margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))

confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

print("Sample Mean:")
print(sample_mean)

print("Confidence interval:")
print(confidence_interval)
```

The sample mean is within the confidence interval
